```python
#  The difference in np.array, np.asarray and np.asanyarray.


#  The difference between Deep copy and shallow copy.
```

np.array()

 np.array() is a function in NumPy that creates a new array from an existing objects (like  lists or other  array_like objects) of from raw data

  it always  returns a new copy of the data. if the inputs is already an ndarray, then it makes a copy of the orignal data by default.


```python
import numpy as np
#creating an array from a list
arr_list= [2,3,4]
array1= np.array(arr_list)
array1
```

np.asarray()

np.asarray() is a function in NumPy that creates an array from the inout, but it tries to avoid making a copy if the input is already an ndarray

if the input is already an ndarray, it won't make a copy of the data ,it will return the original array.

it's useful when you want to ensure that you are working with an ndarray, but you are not sure if the input is already one

```python
import numpy as np
#creating an array from a list
arr_list= [1,2,3]
array2= np.asarray(arr_list)
array2
```

np.asanyarray()

np.asanyarray is similar to np.asarray , but it ensure that the output is an ndarray or a subclass
of ndarray

It's useful when you want to ensure that the output array can be any subclass of ndarray

```python
import numpy as np
#creaing an array from a list
arr_list= [1,2,3]
array3= np.asanyarray(arr_list)
array3
```

```python
print(np.array([1,2,3]))
print(np.asarray([1,2,3]))
print(np.asanyarray([1,2,3]))
```

The difference between Deep copy and shallow copy is significant , especially when dealing with mutable objects like lists or dictionaries.Here's an explanation of each.

Shallow Coyp:

in a shallow copy, a new object is created , but the content itself is not dulipcated. Instead, references to the original objects are copied into the new objects. This means that change made to the original objects will also be reflected in the copied objects, and vice versa.

Shallow copy is a more memory-efficient operation compared to deep copy because it doesn't create new copies of the nested objects.

Shallow copy can be performed using slicing, the copy() method, or the copy module's copy() function in Python.


```python
import copy
#original list
original_list= [1,2,[3,4,5],6]
#shallow copy using slicing
shallow_copy= original_list[:]
#modify the nested list in the shalow copy
shallow_copy[2][1]=30
print("Original list:", original_list)
print("shallow copy:", shallow_copy)
```

```python
# 15. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following
# tasks using Matplotlib:
# a) Plot a histogram of the data with 30 bins.

# b) Overlay a line plot representing the normal distribution's probability density function (PDF).

# c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

# d) Set the title of the plot as 'Histogram with PDF Overlay'.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

data = np.random.normal(0, 1, size=1000)
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')


# b
# Overlay a line plot representing the normal distribution's PDF
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, loc=np.mean(data), scale=np.std(data))
plt.plot(x, p, 'k', linewidth=2, label='PDF')


# c
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
# b
plt.title('Histogram with PDF Overlay')
plt.legend()


plt.show()
```

```python
# 17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
# origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.
```

```python
import seaborn as sns
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Gererate two random arrays
x= np.random.randn(100)
y= np.random.randn(100)

#Determine teh quadrant for each point

def determine_quadrant(x,y):
    if x>0 and y> 0:
        return "Q1"
    elif x<0 and y> 0:
        return "Q2"
    elif x<0 and y<0:
        return "Q3"
    elif x>0 and y<0:
        return "Q4"

#Create a DataFrame
data= pd.DataFrame({'x':x,'y':y})
data['Quadrant']= data.apply(lambda row: determine_quadrant(row['x'], row['y']),axis=1)

#plot using Seaborn
plt.figure(figsize=(10,6))
sns.scatterplot(data, x='x',y='y',hue='Quadrant', palette= 'Set1')

#Add legend ,labels,and title
plt.legend(title='Quadrant')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

#show the plot
plt.show()
```

```python
from bokeh.plotting import figure, show
import numpy as np

# Step 1: Generate data
x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x)

# Step 2: Create a Bokeh figure
p = figure(title='Sine Wave Function', x_axis_label='X-axis', y_axis_label='Y-axis')

# Step 3: Add a line glyph
p.line(x, y, line_width=2)

# Step 4: Add grid lines
p.grid.grid_line_alpha = 0.3

# Step 5: Label axes and set title
p.xaxis.axis_label_text_font_size = '14pt'
p.yaxis.axis_label_text_font_size = '14pt'
p.title.text_font_size = '16pt'

# Step 6: Display the plot
show(p)
```

```python
# 19 Using Bo+eh, generate a bar chart of randomly generated categorical data, color bars based on their
# values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Cateorical
# Bar Chart'.
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import factor_cmap
from bokeh.palettes import Spectral6
import numpy as np
import pandas as pd

# Ensure output is displayed in the notebook
output_notebook()

# Generate random categorical data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(1, 100, size=len(categories))

# Create a DataFrame
data = pd.DataFrame({'categories': categories, 'values': values})

# Create a ColumnDataSource
source = ColumnDataSource(data)

# Create a Bokeh plot
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           x_axis_label='Categories', y_axis_label='Values')

# Add bar renderer with color mapping
p.vbar(x='categories', top='values', width=0.9, source=source,
       line_color='white', fill_color=factor_cmap('categories', palette=Spectral6, factors=categories))

# Add hover tool
hover = HoverTool()
hover.tooltips = [
    ("Category", "@categories"),
    ("Value", "@values"),
]
p.add_tools(hover)

# Show the plot
show(p)
```

```python
# 20. Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as
# 'Simple Line Plot'.
```

```python
import plotly.graph_objs as go
import numpy as np

# Step 1: Generate random data
np.random.seed(0)
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Step 2: Create a Plotly figure
fig = go.Figure()

# Step 3: Add data trace (line plot)
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='sin(x)'))

# Step 4: Label axes and set title
fig.update_layout(
    xaxis_title='X-axis',
    yaxis_title='Y-axis',
    title='Simple Line Plot'
)

# Step 5: Display the plot
fig.show()
```

```python
#21 Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set
# the title as 'Interactive Pie Chart'.
```

```python
import plotly.graph_objs as go
import plotly.offline as pyo
import numpy as np

# Generate random data
np.random.seed(42)  # Seed for reproducibility
labels = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(10, 100, size=len(labels))

# Create a pie chart
trace = go.Pie(
    labels=labels,
    values=values,
    hoverinfo='label+percent',
    textinfo='label+percent'
)

layout = go.Layout(
    title='Interactive Pie Chart'
)

fig = go.Figure(data=[trace], layout=layout)

# Display the plot
pyo.iplot(fig)
```
